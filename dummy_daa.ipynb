```python
#[Problem 1] Creating random numbers
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

mean = [-3, 0]

cov = [[1.0,0.8],[0.8, 1.0]]
x, y = np.random.multivariate_normal(mean, cov, (500,2)).T

plt.plot(x, y, 'o')
plt.axis('equal')
plt.show()
```

```python
#[Problem 2] Visualization by scatter plot
plt.scatter(x, y)
plt.show()
```

```python
#[Problem 3] Visualization by histogram
plt.hist(x)
plt.show()

plt.xlim(-5,5)
plt.hist(y)
plt.show()
```

```python
#[Problem 4] Addition of data
np.random.seed(0)

mean = [0, -3]
cov = [[1.0,0.8],[0.8, 1.0]]
x4, y4 = np.random.multivariate_normal(mean, cov, (500,2)).T


mean = [-3, 0]
cov = [[1.0,0.8],[0.8, 1.0]]
x, y = np.random.multivariate_normal(mean, cov, (500,2)).T

scatter0 = plt.scatter(x, y, color = 'yellow')
scatter1= plt.scatter(x4, y4, color = 'blue')

plt.legend((scatter0,scatter1),
           (0,1),
           loc='lower left')

plt.show()
```

```python
#[Problem 5] Data combination

xcom = np.array(np.concatenate((x, x4), axis = 1))
ycom = np.array(np.concatenate((y, y4), axis = 1))

plt.plot(xcom, ycom, 'o')
plt.show()
```
